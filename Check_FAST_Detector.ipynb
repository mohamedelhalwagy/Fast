```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import copy
%matplotlib inline
```

## Switching image to grey scale 

```python
# Load the image
image = cv2.imread('./data/1.jpg', cv2.IMREAD_GRAYSCALE)

_ = plt.imshow(image)
```

## Initiating the model with defaults 

```python
fast = cv2.FastFeatureDetector.create()
keypoints = fast.detect(image)

print("Number of detected keypoints: ", len(keypoints))
image_copy = cv2.drawKeypoints(image, keypoints, image)
_ = plt.imshow(image_copy)
```
